## Visualization & Explanatory Analysis

### Pertanyaan 1:

```python
plt.figure(figsize=(12, 5))
sns.lineplot(x=df['month'], y=df['PM2.5'], hue=df['year'], marker='o')

plt.xlabel('Bulan')
plt.ylabel('Rata-rata PM2.5')
plt.title('Tren Polusi PM2.5 Sepanjang Tahun')
plt.legend(title='Tahun')
plt.show()
```

### Pertanyaan 2:

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.scatterplot(x=df['WSPM'], y=df['PM2.5'], alpha=0.5)
plt.xlabel('Kecepatan Angin (WSPM)')
plt.ylabel('Konsentrasi PM2.5')
plt.title('Hubungan Kecepatan Angin dan Polusi PM2.5')
plt.show()

correlation = df[['WSPM', 'PM2.5']].corr()
print(correlation)
```

**Insight:**
- Pertanyaan 1
Polusi udara cenderung lebih tinggi saat tekanan udara rendah, Hal ini mungkin karena udara stagnan dan tidak ada angin yang membantu menyebarkan polutan.

- Pertanyaan 2
PM2.5 dan PM10 memiliki korelasi positif yang cukup kuat → Artinya, jika PM2.5 naik, PM10 juga cenderung ikut naik.
